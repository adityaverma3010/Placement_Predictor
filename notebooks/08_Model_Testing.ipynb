```python
import logging
import pandas as pd

import src.logger

from src.predict import PredictionPipeline
from src.utils import load_object
from config.config import MODEL_PATH
```

```python
logging.info("=" * 60)
logging.info("LOADING TRAINED MODEL")
logging.info("=" * 60)

pipeline = load_object(MODEL_PATH)

print("Model Loaded Successfully")
```

```python
predictor = PredictionPipeline(pipeline)
```

```python
sample = {
    "IQ": 120,
    "Prev_Sem_Result": 8.5,
    "CGPA": 8.6,
    "Academic_Performance": 9,
    "Internship_Experience": 1,
    "Extra_Curricular_Score": 8,
    "Communication_Skills": 9,
    "Projects_Completed": 4
}
```

```python
prediction, probability = predictor.predict(sample)
```

```python
print("Prediction :", prediction)

if prediction == 1:
    print("Student is Likely to be Placed")
else:
    print("Student is Not Likely to be Placed")

print(f"Placement Probability : {probability[1]*100:.2f}%")
print(f"Not Placement Probability : {probability[0]*100:.2f}%")
```
